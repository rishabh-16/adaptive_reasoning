```python
import matplotlib.pyplot as plt

# Data extracted from the screenshot
N_values = [1, 4, 16]

data = {
    4:  [0.254, 0.270, 0.337],
    8:  [0.476, 0.453, 0.554],
    16: [0.336, 0.800, 1.000],
}

# Plot
plt.figure()
for wbits, costs in data.items():
    plt.plot(N_values, costs, marker='o', label=f'WBits = {wbits}')

plt.xlabel('N (Best-of-N)')
plt.ylabel('Normalized Cost = cost / max-cost')
plt.title('Normalized Cost vs N (Best-of-N) for Different WBits')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Normalized Cost (x) and Accuracy (y) for each WBits (after excluding 2 and 3bit)
data = {
    4:  [(0.254, 16.667), (0.270, 17.778), (0.337, 16.667)],
    8:  [(0.476, 16.667), (0.453, 15.556), (0.554, 17.778)],
    16: [(0.336, 18.889), (0.800, 17.778), (1.000, 17.778)],
}

# Sort each curve in ascending cost so lines connect from low  high cost
for k in data:
    data[k] = sorted(data[k], key=lambda t: t[0])

plt.figure()
for wbits, points in data.items():
    xs, ys = zip(*points)
    plt.plot(xs, ys, marker='o', label=f'WBits = {wbits}')

plt.xlabel('Normalized Cost = cost / max-cost')
plt.ylabel('Accuracy (%)')
plt.title('Accuracy vs Normalized Cost for Different WBits')
plt.legend()
plt.grid(True)
plt.show()
```
