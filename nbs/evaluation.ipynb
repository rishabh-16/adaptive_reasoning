```python
def calculate_pass_at_k(list_of_scores, k=8):
    """Calculate pass@k metric for a list of problem scores"""
    pass_at_k = 0
    for score in list_of_scores:
        if any(score[:k]):
            pass_at_k += 1
    return pass_at_k / len(list_of_scores)

def calculate_acc_at_k(list_of_scores, k=8):
    acc_at_k = 0
    for score in list_of_scores:
        acc_at_k += np.sum(score[:k])
    return acc_at_k / (len(list_of_scores)*k)
```

```python
import json
from collections import defaultdict

# Define thinking budget and top_k values to test
thinking_budgets = [4096, 8192, 16384]
top_k_values = [8, 12, 16, 20]

def calculate_pass_at_k(problem_scores, k=8):
    """Calculate pass@k metric for a list of problem scores"""
    if len(problem_scores) < k:
        return sum(problem_scores) / len(problem_scores) if problem_scores else 0
    
    # Check if at least one of the k attempts was correct
    return 1 if any(problem_scores[:k]) else 0

results_pass_at_8 = {}

for tb in thinking_budgets:
    results_pass_at_8[tb] = {}
    tokens = tb + 4096
    
    for top_k in top_k_values:
        # Group scores by problem_id
        problem_scores_dict = defaultdict(list)
        
        for seed in range(8):
            filepath = f"/home/rishabhtiwari/adaptive_reasoning/experiments/03_evaluation/outputs/base-1235229_checkpoint1104_max_tokens_per_call{tokens}_thinking_budget{tb}/aime25/test_qwen25-math-cot_-1_seed{seed}_t0.6_top_k{top_k}_s0_e-1.jsonl"
            
            try:
                with open(filepath, 'r') as f:
                    for line in f:
                        data = json.loads(line.strip())
                        print(data['idx'])
                        raise Exception("stop")
                        problem_id = data.get('idx', 'unknown')
                        score = 1 if data['score'] == [True] else 0
                        problem_scores_dict[problem_id].append(score)
            except FileNotFoundError:
                print(f"File not found: {filepath}")
                continue
        
        if problem_scores_dict:
            # Calculate pass@8 for each problem
            pass_at_8_scores = []
            for problem_id, scores in problem_scores_dict.items():
                pass_at_8_scores.append(calculate_pass_at_k(scores, k=8))
            
            overall_pass_at_8 = sum(pass_at_8_scores) / len(pass_at_8_scores)
            results_pass_at_8[tb][top_k] = {
                'pass_at_8': overall_pass_at_8,
                'total_problems': len(pass_at_8_scores),
                'total_samples': sum(len(scores) for scores in problem_scores_dict.values())
            }
            print(f"Thinking Budget {tb}, Top-K {top_k}: Pass@8 = {overall_pass_at_8:.2f}% (problems={len(pass_at_8_scores)}, samples={sum(len(scores) for scores in problem_scores_dict.values())})")

print("\nPass@8 Summary:")
for tb in thinking_budgets:
    print(f"\nThinking Budget {tb}:")
    for top_k in top_k_values:
        if top_k in results_pass_at_8[tb]:
            result = results_pass_at_8[tb][top_k]
            print(f"  Top-K {top_k}: Pass@8 = {result['pass_at_8']:.2f}%, problems={result['total_problems']}, samples={result['total_samples']}")
```
